# Prepocessing Dataset

```python
# import library 
import pandas as pd
```

```python
# import dataset
df = pd.read_csv('skincare.csv')
```

```python
# eksplorasi awal
print("Shape:", df.shape)
print("\nKolom:", df.columns.tolist())
print("\nSample data:")
print(df.head())
```

```python
# ── CEK TIPE DATA & INFO UMUM ────────────────────────────────
print("\nInfo dataset:")
print(df.info())
```

```python
# ── CEK MISSING VALUES ───────────────────────────────────────
print("\nMissing values per kolom:")
print(df.isnull().sum())
print("\nPersentase missing:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# ── CEK TIAP KOLOM ───────────────────────────────────────────
# Dari sini kamu bisa lihat sendiri kolom mana yang relevan

print("\nUnique values kolom 'type':")
print(df['type'].value_counts().to_string())

print("\nUnique values kolom 'country':")
print(df['country'].value_counts().to_string())

print("\nContoh isi kolom 'afterUse':")
print(df['afterUse'].dropna().head(5).to_string())

print("\nContoh isi kolom 'ingridients':")
print(df['ingridients'].dropna().head(3).to_string())
```

```python
# ── 1. DROP KOLOM TIDAK RELEVAN ──────────────────────────────
df = df.drop(columns=['country'])
```

```python
# ── 2. DROP MISSING VALUES ───────────────────────────────────
print("Shape sebelum drop missing:", df.shape)
df = df.dropna(subset=['type', 'afterUse', 'ingridients'])
print("Shape setelah drop missing:", df.shape)
```

```python
# ── 3. FILTER TYPE + GABUNGKAN MOISTURIZER ───────────────────
skincare_types = [
    'Serum', 'General Moisturizer', 'Face Cleanser',
    'Sunscreen', 'Toner', 'Exfoliator', 'Makeup Remover',
    'Day Moisturizer', 'Night Moisturizer'
]

df = df[df['type'].isin(skincare_types)].copy()

# Gabungkan semua moisturizer jadi satu label
df['type'] = df['type'].replace({
    'General Moisturizer': 'Moisturizer',
    'Day Moisturizer': 'Moisturizer',
    'Night Moisturizer': 'Moisturizer'
})

print("\nShape setelah filter:", df.shape)
print("\nType yang tersisa:")
print(df['type'].value_counts())
```

```python
# ── CEK sebelum encoding ─────────────────────────────────────
from collections import Counter

all_tags = []
for x in df['afterUse'].dropna():
    all_tags.extend([tag.strip() for tag in x.split(',')])

tag_counts = Counter(all_tags)
print("Semua tag unik di afterUse:")
for tag, count in sorted(tag_counts.items(), key=lambda x: -x[1]):
    print(f"{count:5d}  {tag}")
print("\nTotal tag unik:", len(tag_counts))
```

```python
# ── CEK mapping dulu ─────────────────────────────────────────
tag_mapping = {
    'Good For Oily Skin'    : 'Oily',
    'May Worsen Oily Skin'  : 'Worsen_Oily',
    'Acne Fighting'         : 'Acne_Fighting',
    'Acne Trigger'          : 'Acne_Trigger',
    'Redness Reducing'      : 'Redness',
    'Reduces Large Pores'   : 'Large_Pores',
    'Reduces Irritation'    : 'Reduces_Irritation',
    'Irritating'            : 'Irritating',
    'Drying'                : 'Drying',
    'Skin Texture'          : 'Skin_Texture',
    'Dark Spots'            : 'Dark_Spots',
    'Scar Healing'          : 'Scar',
    'Anti-Aging'            : 'Anti_Aging',
    'Brightening'           : 'Brightening',
    'Hydrating'             : 'Hydrating',
    'Eczema'                : 'Eczema',
    'Rosacea'               : 'Rosacea',
}
```

```python
# Cek apakah semua tag di dataset ter-cover di mapping
all_tags = set()
for x in df['afterUse'].dropna():
    all_tags.update([tag.strip() for tag in x.split(',')])

print("Tag di dataset tapi tidak ada di mapping:")
print(all_tags - set(tag_mapping.keys()))

print("\nTag di mapping tapi tidak ada di dataset:")
print(set(tag_mapping.keys()) - all_tags)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# ── APPLY MAPPING + ENCODING ──────────────────────────────────
df['afterUse_list'] = df['afterUse'].apply(
    lambda x: [tag_mapping[tag.strip()] for tag in x.split(',')]
)

mlb = MultiLabelBinarizer()
afteruse_encoded = pd.DataFrame(
    mlb.fit_transform(df['afterUse_list']),
    columns=mlb.classes_,
    index=df.index
)

# ── GABUNGKAN ─────────────────────────────────────────────────
df_final = pd.concat(
    [df[['brand', 'name', 'type']], afteruse_encoded],
    axis=1
)

# ── CEK HASIL ─────────────────────────────────────────────────
print("Shape:", df_final.shape)
print("\nKolom binary:", afteruse_encoded.columns.tolist())
print("\nSample:")
print(df_final.head(3).to_string())
```

```python
# ── CEK type yang tersisa setelah preprocessing ───────────────
print("Type yang ada:")
print(df_final['type'].value_counts())
```

```python
# ── PISAH PER TYPE ────────────────────────────────────────────
df_serum = df_final[df_final['type'] == 'Serum'].copy()
df_moisturizer = df_final[df_final['type'] == 'Moisturizer'].copy()
df_cleanser = df_final[df_final['type'] == 'Face Cleanser'].copy()
df_sunscreen = df_final[df_final['type'] == 'Sunscreen'].copy()
df_toner = df_final[df_final['type'] == 'Toner'].copy()
df_exfoliator = df_final[df_final['type'] == 'Exfoliator'].copy()
df_makeup_remover = df_final[df_final['type'] == 'Makeup Remover'].copy()

# ── CEK ───────────────────────────────────────────────────────
print("Serum:", df_serum.shape)
print("Moisturizer:", df_moisturizer.shape)
print("Face Cleanser:", df_cleanser.shape)
print("Sunscreen:", df_sunscreen.shape)
print("Toner:", df_toner.shape)
print("Exfoliator:", df_exfoliator.shape)
print("Makeup Remover:", df_makeup_remover.shape)
```

```python
# ── GABUNGKAN ULANG + TAMBAH INGRIDIENTS ──────────────────────
df_final = pd.concat(
    [df[['brand', 'name', 'type', 'ingridients']], afteruse_encoded],
    axis=1
)

# ── CEK ───────────────────────────────────────────────────────
print("Shape:", df_final.shape)
print("\nKolom:", df_final.columns.tolist())
print("\nSample:")
print(df_final[['brand', 'name', 'type', 'ingridients']].head(3).to_string())
```

```python
# ── SIMPAN ────────────────────────────────────────────────────
df_final.to_csv('skincare_preprocessed.csv', index=False)
print("\nTersimpan!")
```
